```python
import pandas as pd

df = pd.read_csv("healthcare_dataset.csv")
df.columns = df.columns.str.strip()
df.describe
print(df.columns)
print(df.head())
```

# **SCATTER PLOT**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x='Height (cm)', y='Weight (kg)', hue='Gender')
plt.title('Height vs Weight by Gender')
plt.show()
```

# **BAR PLOT**

```python
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='Blood_Type', y='BMI')
plt.title('Average BMI by Blood Type')
plt.show()
```

# **BOX PLOT**

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x='Gender', y='Heart_Rate (bpm)')
plt.title('Heart Rate Distribution by Gender')
plt.show()
```

# **PIE CHART**

```python
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

# **LINE CHART**

```python
sorted_df = df.sort_values('Age')
plt.figure(figsize=(6, 4))
plt.plot(sorted_df['Age'], sorted_df['BMI'], marker='o')
plt.title('Age vs BMI')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.grid(True)
plt.show()
```
